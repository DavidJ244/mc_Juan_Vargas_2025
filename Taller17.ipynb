```python
import numpy as np
import matplotlib.pyplot as plt

# Datos del taller
x = np.array([0, 1, 2, 3, 4, 5, 6, 7])
y = np.array([7.5, 5.5, 6.5, 3.5, 4.5, 3, 2.5, 1])
n = len(x)

#promedios
x_prom = np.mean(x)
y_prom = np.mean(y)

#coeficientes a1 y a0
a1 = (n * np.sum(x*y) - np.sum(x)*np.sum(y)) / (n*np.sum(x**2) - (np.sum(x))**2)
a0 = y_prom - a1 * x_prom

#valores predichos por la recta
y_ajuste = a0 + a1 * x

#ecuaci√≥n
print(f"Ecuaci√≥n ajustada: y = {a0:.4f} + {a1:.4f}x")


#puntos y l√≠nea ajustada
plt.figure(figsize=(8, 5))
plt.scatter(x, y, color='blue', label='Datos reales')
plt.plot(x, y_ajuste, color='red', linestyle='--', label='L√≠nea ajustada')
plt.title('Regresi√≥n Lineal por M√≠nimos Cuadrados')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)
plt.show()
```
