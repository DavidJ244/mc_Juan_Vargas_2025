```python
from itertools import product

# Definir las operaciones booleanas usando 0 y 1
def boolean_and(x, y):
    return x & y  # AND l贸gico

def boolean_or(x, y):
    return x | y  # OR l贸gico

def boolean_not(x):
    return 1 - x  # NOT l贸gico

# Generar todas las combinaciones posibles de valores
valores_3 = list(product([0, 1], repeat=3))  # Para leyes con X, Y, Z
valores_2 = list(product([0, 1], repeat=2))  # Para leyes con X, Y

# Funci贸n para imprimir tablas de verdad
def imprimir_tabla(headers, datos):
    print("\n" + " | ".join(headers))
    print("-" * (len(headers) * 7))
    for fila in datos:
        print(" | ".join(str(val) for val in fila))
    print("\n")

# 1锔 Ley Asociativa (OR)
print(" Ley Asociativa (OR): X or (Y or Z) == (X or Y) or Z")
tabla_asociativa = []
for x, y, z in valores_3:
    col1 = boolean_or(y, z)
    col2 = boolean_or(x, col1)
    col3 = boolean_or(x, y)
    col4 = boolean_or(col3, z)
    tabla_asociativa.append([x, y, z, col1, col2, col3, col4])
imprimir_tabla(["X", "Y", "Z", "Y or Z", "X or (Y or Z)", "X or Y", "(X or Y) or Z"], tabla_asociativa)

# 2锔 Ley Conmutativa (OR)
print(" Ley Conmutativa (OR): X or Y == Y or X")
tabla_conmutativa = []
for x, y in valores_2:
    tabla_conmutativa.append([x, y, boolean_or(x, y), boolean_or(y, x)])
imprimir_tabla(["X", "Y", "X or Y", "Y or X"], tabla_conmutativa)

# 3锔 Ley Distributiva: X and (Y or Z) == (X and Y) or (X and Z)
print(" Ley Distributiva: X and (Y or Z) == (X and Y) or (X and Z)")
tabla_distributiva = []
for x, y, z in valores_3:
    col1 = boolean_or(y, z)
    col2 = boolean_and(x, col1)
    col3 = boolean_and(x, y)
    col4 = boolean_and(x, z)
    col5 = boolean_or(col3, col4)
    tabla_distributiva.append([x, y, z, col1, col2, col3, col4, col5])
imprimir_tabla(["X", "Y", "Z", "Y or Z", "X and (Y or Z)", "X and Y", "X and Z", "(X and Y) or (X and Z)"], tabla_distributiva)

# 4锔 Ley de Identidad: X or 0 == X, X and 1 == X
print(" Ley de Identidad: X or 0 == X, X and 1 == X")
tabla_identidad = []
for x in [0, 1]:
    tabla_identidad.append([x, boolean_or(x, 0), boolean_and(x, 1)])
imprimir_tabla(["X", "X or 0", "X and 1"], tabla_identidad)

# 5锔 Ley de Complemento: X or not X == 1, X and not X == 0
print(" Ley de Complemento: X or not X == 1, X and not X == 0")
tabla_complemento = []
for x in [0, 1]:
    col1 = boolean_not(x)
    col2 = boolean_or(x, col1)
    col3 = boolean_and(x, col1)
    tabla_complemento.append([x, col1, col2, col3])
imprimir_tabla(["X", "not X", "X or not X", "X and not X"], tabla_complemento)

```
