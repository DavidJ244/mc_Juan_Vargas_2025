```python
import numpy as np
import matplotlib.pyplot as plt

# Datos
x = np.array([0, 1, 2, 3, 4, 5, 6])
y = np.array([0, 0.5, 2, 3.5, 4.5, 9, 13.5])
n = len(x)

# Promedios
x_prom = np.mean(x)
y_prom = np.mean(y)

# Coeficientes de la recta
a1 = (n*np.sum(x*y) - np.sum(x)*np.sum(y)) / (n*np.sum(x**2) - (np.sum(x))**2)
a0 = y_prom - a1 * x_prom

# Línea ajustada
y_ajustada = a0 + a1*x

# Cálculo de Sr (residuos al cuadrado)
Sr = np.sum((y - y_ajustada)**2)

# Cálculo de St (diferencia respecto a la media)
St = np.sum((y - y_prom)**2)

# Desviación estándar sy
sy = np.sqrt(St / (n - 1))

# Error estándar de estimación
s_yx = np.sqrt(Sr / (n - 2))

# Coeficientes de ajuste
r2 = (St - Sr) / St
r = np.sqrt(r2) * 100  # En porcentaje

# Resultados
print(f"Ecuación ajustada: y = {a0:.4f} + {a1:.4f}x")
print(f"Desviación estándar (sy): {sy:.4f}")
print(f"Error estándar de estimación (sy|x): {s_yx:.4f}")
print(f"Coeficiente de determinación (r²): {r2:.4f}")
print(f"Coeficiente de correlación (r): {r:.2f}%")

# Gráfica
plt.figure(figsize=(8, 5))
plt.scatter(x, y, label="Datos reales", color="blue")
plt.plot(x, y_ajustada, label="Regresión lineal", color="red", linestyle="--")
plt.title("Regresión Lineal y Métricas de Ajuste")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.legend()
plt.show()
```
