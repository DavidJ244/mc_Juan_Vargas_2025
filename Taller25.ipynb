```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import CubicSpline

x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 5, 2.5, 4, -1.6, 2])

# Interpolación con Lagrange
def lagrange_interp(x_data, y_data, x_eval):
    total = 0
    n = len(x_data)
    for i in range(n):
        L = 1
        for j in range(n):
            if i != j:
                L *= (x_eval - x_data[j]) / (x_data[i] - x_data[j])
        total += L * y_data[i]
    return total

# Estimaciones
x_objetivo = 3.55
f_lagrange = lagrange_interp(x, y, x_objetivo)
f_spline = CubicSpline(x, y)(x_objetivo)

print(f"Lagrange: f(3.55) ≈ {f_lagrange:.4f}")
print(f"Trazador cúbico: f(3.55) ≈ {f_spline:.4f}")

# Gráfica
x_smooth = np.linspace(0, 5, 200)
plt.plot(x, y, 'o', label='Datos')
plt.plot(x_smooth, CubicSpline(x, y)(x_smooth), label='Trazador cúbico')
plt.axvline(x=3.55, linestyle='--', color='gray', label='x = 3.55')
plt.title("Interpolación con trazadores cúbicos")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid()
plt.show()
```
