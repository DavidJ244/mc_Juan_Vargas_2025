```python
import numpy as np
import matplotlib.pyplot as plt

# Datos
x = np.array([1, 2, 3, 4, 5, 6])
y = np.array([1.5, 2.5, 3.5, 4.5, 6.5, 9])
ln_y = np.log(y)  # Transformación logarítmica

# Regresión lineal sobre (x, ln(y))
n = len(x)
x_prom = np.mean(x)
lny_prom = np.mean(ln_y)

# Cálculo de los coeficientes
beta = (n*np.sum(x*ln_y) - np.sum(x)*np.sum(ln_y)) / (n*np.sum(x**2) - (np.sum(x))**2)
ln_alpha = lny_prom - beta * x_prom
alpha = np.exp(ln_alpha)

# Modelo exponencial: y = α * e^(βx)
y_modelo = alpha * np.exp(beta * x)

# Cálculo de Sr y St
Sr = np.sum((y - y_modelo)**2)
St = np.sum((y - np.mean(y))**2)

# Errores y coeficientes
syx = np.sqrt(Sr / (n - 2))
r2 = (St - Sr) / St
r = np.sqrt(r2) * 100  # En porcentaje

# Resultados
print(f"Modelo ajustado: y = {alpha:.4f} * e^({beta:.4f}x)")
print(f"Error estándar de estimación: {syx:.4f}")
print(f"Coeficiente de determinación (r²): {r2:.4f}")
print(f"Coeficiente de correlación (r): {r:.2f}%")

# Gráfica
plt.figure(figsize=(8, 5))
plt.scatter(x, y, color='blue', label='Datos reales')
plt.plot(x, y_modelo, color='red', linestyle='--', label='Modelo exponencial ajustado')
plt.title('Regresión Exponencial por Mínimos Cuadrados')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```
