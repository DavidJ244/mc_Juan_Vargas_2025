```python
import random

class Nodo:
    def __init__(self, valor):
        self.valor = valor
        self.izq = None
        self.der = None

class ArbolBinario:
    def __init__(self):
        self.raiz = None

    def insertar(self, valor):
        self.raiz = self._insertar(self.raiz, valor)

    def _insertar(self, nodo, valor):
        if nodo is None:
            return Nodo(valor)
        if valor < nodo.valor:
            nodo.izq = self._insertar(nodo.izq, valor)
        elif valor > nodo.valor:
            nodo.der = self._insertar(nodo.der, valor)
        return nodo

    def buscar(self, valor):
        return self._buscar(self.raiz, valor)

    def _buscar(self, nodo, valor):
        if nodo is None:
            return False
        if valor == nodo.valor:
            return True
        elif valor < nodo.valor:
            return self._buscar(nodo.izq, valor)
        else:
            return self._buscar(nodo.der, valor)

    def imprimir_ordenado(self):
        self._imprimir_ordenado(self.raiz)
        print()

    def _imprimir_ordenado(self, nodo):
        if nodo:
            self._imprimir_ordenado(nodo.izq)
            print(nodo.valor, end=' ')
            self._imprimir_ordenado(nodo.der)

# Crear árbol con 20 números aleatorios
valores = random.sample(range(1, 101), 20)
arbol = ArbolBinario()
for val in valores:
    arbol.insertar(val)

# Menú interactivo
while True:
    print("\n1. Buscar número")
    print("2. Mostrar en orden")
    print("3. Salir")
    opcion = input("Seleccione una opción: ")

    if opcion == '1':
        numero = int(input("Ingrese número a buscar: "))
        if arbol.buscar(numero):
            print("El número está en el árbol.")
        else:
            print("El número NO está en el árbol.")
    elif opcion == '2':
        print("Elementos en orden ascendente:")
        arbol.imprimir_ordenado()
    elif opcion == '3':
        break
    else:
        print("Opción inválida.")

```
