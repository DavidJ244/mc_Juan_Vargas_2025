```python
import numpy as np
import matplotlib.pyplot as plt

# Datos
x = np.array([0, 2, 4, 6, 8, 10, 12], dtype=float)
y = np.array([7.5, 1.8, -1, -1.8, -1.2, 2.2, 7.2], dtype=float)
n = len(x)

# Ajuste de polinomio de grado 2
coeficientes = np.polyfit(x, y, 2)  # [a2, a1, a0]
a2, a1, a0 = coeficientes
y_ajustada = a0 + a1*x + a2*x**2

# Cálculo de Sr y St
Sr = np.sum((y - y_ajustada)**2)
St = np.sum((y - np.mean(y))**2)

# Coeficiente de determinación y correlación
r2 = (St - Sr) / St
r = np.sqrt(r2) * 100  # porcentaje

# Mostrar resultados
print(f"Polinomio ajustado: y = {a0:.4f} + {a1:.4f}x + {a2:.4f}x²")
print(f"Coeficiente de determinación r² = {r2:.4f}")
print(f"Coeficiente de correlación r = {r:.2f}%")

# Graficar puntos y curva
x_graf = np.linspace(min(x), max(x), 200)
y_graf = a0 + a1*x_graf + a2*x_graf**2

plt.figure(figsize=(8, 5))
plt.scatter(x, y, color='blue', label='Datos reales')
plt.plot(x_graf, y_graf, color='red', linestyle='--', label='Polinomio ajustado')
plt.title("Regresión Polinomial (grado 2)")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()

```
